# Here we proceed to merge every df and operate some operation 

```python
import pandas as pd 
```

```python
a = pd.read_csv('df_Obama2012.csv')
b = pd.read_csv('df_trump2016.csv')
c = pd.read_csv('df_trump.csv')
```

```python
a['Mandat'] = 'Obama'
b['Mandat'] = 'Trump1'
c['Mandat'] = 'Trump2'
```

```python
print(type(a))
print(type(b))
print(type(c))
```

```python
print(a.columns)
print(b.columns)
print(c.columns)
```

```python
print(a.dtypes)
print(b.dtypes)
print(c.dtypes)
```

```python
for df in [a, b, c]:
    # We make sure that date is in datetime
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    # Columns to convert 
    numeric_columns = ['Opening', 'Highest', 'Lowest', 'Fermer', 'Cloture ajusté', 'Volume']
    for col in numeric_columns:
        df[col] = (
            df[col]
            .str.replace(" ", "", regex=False)  # Delete the spaces 
            .str.replace(",", ".", regex=False)  # Replace the commas by dots
        )
        # Convertir en float après nettoyage
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(a.dtypes)
print(b.dtypes)
print(c.dtypes)
```

```python
df_combined = pd.concat([a, b], ignore_index=True)
df_combined = pd.concat([df_combined, c], ignore_index=True)
```

```python
df_combined
```

```python
# Iintify the column in floats 
float_columns = df_combined.select_dtypes(include=['float']).columns

# Apply the division by 100 for every indexes exepts VIX
for col in float_columns:
    df_combined.loc[df_combined['Indice'] != 'VIX', col] = df_combined.loc[df_combined['Indice'] != 'VIX', col] / 100
```
